```python
def run_VR_model_1(model_data,old,mip):
    solver = SolverFactory("gurobi")
    #solver = SolverFactory("xpress")
    model = make_VR_model()
    instance = model.create_instance(model_data)
    for i in old.N:
        instance.S_R[i] = old.S_R[i].value
        instance.S_W[i] = old.S_W[i].value
        instance.S_Q[i] = old.S_Q[i].value
        for bus in old.B:
            instance.T[i,bus] = old.T[i,bus].value
            instance.Q_track[i,bus] = old.Q_track[i,bus].value
            for j in old.N:
                instance.x[i,j,bus] = old.x[i,j,bus].value
                instance.z[i,j,bus] = old.z[i,j,bus].value


    solver.options['mipgap'] = str(mip)
    #solver.options['MIPFocus'] = '3'
    #solver.options['TimeLimit'] = '30' 
    #solver.options['ConcurrentMIP'] = '3'
    instance.preprocess()
    results = solver.solve(instance,tee=True,warmstart=True)
    instance.solutions.load_from(results)
    #instance.display() # shows everything from solution
    
    return instance, results
```
